```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
matplotlib.style.use('ggplot')
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
dt = pd.read_csv('/Users/yaserkeneshloo/Downloads/green_tripdata_2017-06.csv',sep=',',header=0)
dt = dt[np.isfinite(dt['trip_distance'])]
fl = dt[(dt['trip_distance']>0)]
tripdistance = fl['trip_distance'].values
```

```python
nums, bins = np.histogram(tripdistance,bins=1000)
plt.figure();
plt.loglog(bins[1:],nums)
plt.axvline(x=6.7, color='k', linestyle='--')
plt.ylabel('ln(Frequency of x)',fontsize=16);
plt.xlabel('ln(x)',fontsize=16)
plt.title('NYC Green Taxi Trip Distance')
plt.show()
```
